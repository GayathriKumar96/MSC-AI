```python
import  pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('http://raptor.kent.ac.uk/~ds756/Data/kc_house_data_small.csv')
df
```

```python
"""Task 1 (10 marks) 
Create a feature in your dataframe, called “unit_price”, that contains price per square foot for each sale.
Use the “sqft_living” feature."""

pd.set_option('display.float_format', '{:.2f}'.format)

df["unit_price"] = df["price"]/df["sqft_living"]
df
```

```python
"""Task 2 (20 marks)
Create a dataframe that consists of two columns.  The columns are “zipcode” and the average price per 
“sqft_living” in the corresponding zip code.  Each zip code will have its own row, and the second column 
will be the mean price of all the sales in said zip code. """

task2_df = df[["zipcode","sqft_living"]].groupby(["zipcode"]).mean()
task2_df
```

```python
"""Task 3 (20 marks) 
Create a dataframe with two columns: bedrooms and mean price.  Each unique count of bedrooms will have its own 
row, and the second column will be the mean price for properties with that many bedrooms.  Use a bar plot to
display your dataframe.  Give each bar its own colour. """

task3_df = df[["bedrooms","price"]].groupby(["bedrooms"]).mean().rename({"price":"mean price"},axis='columns')
task3_df = task3_df.reset_index()
bar_graph = plt.bar(task3_df["bedrooms"], task3_df["mean price"])
colors = ['b','g','r','c','m','y','k','pink','aquamarine','mediumseagreen']
for i in range(len(bar_graph)):
    bar_graph[i].set_color(colors[i])
plt.show()
```

```python
"""Task 4 (20 marks) 
Graph “long” versus “lat”.  Colour the waterfront properties red, and all others  as blue. 
Title the plot and label the axis appropriately. """
plt.xlabel('Lat')
plt.ylabel('Long')
plt.title ("Long vs Lat Graph")

plt.plot(df['lat'], df['long'], "ob")

waterfront_df = df.loc[df['waterfront']==1]
plt.plot(waterfront_df['lat'], waterfront_df['long'], "ro")
plt.show()
```

```python
"""Task 5 (30 marks)
Create a time-series from the dataframe (set the index to the date). Plot the daily mean price. 
On the same plot, draw (overlay) a line from the smallest daily average to the highest daily average. 
The line should appear on top of the timeseries. """


task5_df = df[["date","price"]].groupby(["date"]).mean()
task5_df.index = pd.to_datetime(task5_df.index)

plt.ylabel('Mean Price')
plt.xlabel('Date')
plt.title ("Daily Mean Price Graph")
plt.plot(task5_df.index, task5_df['price'], 'ob')

x_values = (task5_df[['price']].idxmin()[0] ,task5_df[['price']].idxmax()[0])
y_values = (task5_df['price'].min(),task5_df['price'].max())
plt.plot(x_values, y_values, 'ro',linestyle="-")
plt.show()
```
