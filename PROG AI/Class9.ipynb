```python
""" * The mode is a measure of centre of a set (as are mean and median). It 
is the element with the highest frequency. 
mode ([4, 4, 1, 5, 5, 5, 7]) = 5 # There are more instances of 5 
Implement a func/on that computes the mode of a column. Name it 
ComputeMode. It will be invoked as ComputeMode 
(your_data_frame[“column name”]). You will have to do something to 
the column before you call your func/on to ensure that it runs in linear 
/me. 
Demonstrate correctness by using it to compute the mode of “Age”"""

def ComputeMode(col):
    value_count={}
    for elem in col:
        if elem not in value_count.keys():
            value_count[elem] = 0
        value_count[elem] += 1
    max_value = max(value_count.values())
    for k in value_count.keys():
        if value_count[k] == max_value:
            return k

print(ComputeMode(df['Age']))
```
